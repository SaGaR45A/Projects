```python
from tkinter import *



root =Tk()
root.configure(background="black")
root.geometry("500x650")
root.title("Calculator")

def click(event):
    global scvalue
    text=event.widget.cget("text")
    print(text)
    if text=="=":
        if scvalue.get().isdigit():
            value= int(scvalue.get())
        else:
             try:
                value = eval(screen.get())
             except Exception as e:
                print(e)
                value = "Error"
        scvalue.set(value)
        screen.update()
        
    elif text=="c":
        scvalue.set("")
        screen.update()
    else:
        scvalue.set(scvalue.get()+text)
        screen.update()
# gui logic
scvalue = StringVar()
scvalue.set("")
screen = Entry(root,textvar=scvalue,width=50,font="40")
screen.pack(fill=X,ipady=8,pady=10,padx=10)


f = Frame(root,bg="white")

b=Button(f,text="c",font="lucida 25 bold")
b.pack(side=LEFT,ipadx=60,padx=18,pady=12)
b.bind("<Button-1>",click)

f.pack()


f = Frame(root,bg="white")

b=Button(f,text="9",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=18,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="8",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=18,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="7",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=18,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="+",font="lucida 25 bold")
b.pack(side=LEFT,padx=20,pady=12)
b.bind("<Button-1>",click)

f.pack()
f = Frame(root,bg="white")


b=Button(f,text="6",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="5",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="4",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="-",font="lucida 25 bold")
b.pack(side=LEFT,padx=20,pady=22)
b.bind("<Button-1>",click)


f.pack()

f = Frame(root,bg="white")

b=Button(f,text="3",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="2",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="1",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="/",padx=5,font="lucida 25 bold")
b.pack(side=LEFT,padx=20,pady=12)
b.bind("<Button-1>",click)

f.pack()

f = Frame(root,bg="white")
b=Button(f,text="0",padx=28,pady=22,font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="*",font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="%",font="lucida 25 bold")
b.pack(side=LEFT,padx=19,pady=12)
b.bind("<Button-1>",click)

b=Button(f,text="=",font="lucida 25 bold")
b.pack(side=LEFT,ipadx=12,padx=19,pady=12)
b.bind("<Button-1>",click)

f.pack()



root.mainloop()
```
